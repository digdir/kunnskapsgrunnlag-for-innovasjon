```python
import numpy as np
import pandas as pd

from src.embedding_model import embed
from src.plot_utils import plot_directorates_yearly_score
from src.utils import (
    get_directorates,
    populate_everything,
)
```

```python
df: pd.DataFrame = pd.read_pickle("df_with_embeddings_float32_filtered50.pkl")
```

```python
inn_texts = [
    "agile tilnærminger og smidig arbeidsmetodikk.",
    "bruk av smidige metodikk",
    "smidig metodikk",
    "ved å bruke smidige teknikker",
    "hyppig iterasjon",
    "smidig",
    "smidig utvikling",
    "kontinuerlig tilpasning og forbedring av arbeidsprosesser for å møte endringer og utfordringer på en smidig måte",
    "smidige arbeidsmetoder",
    "agile prinsipper",
    """Anvende smidige arbeidsmetoder. Smidige prinsipper og teknikker vil bidra til økt fleksibilitet, raskere tilpasningsevne og bedre leveranse av resultater som samsvarer med departementets målsettinger og forventninger. Smidig arbeidsmetodikk er en effektiv tilnærming for å håndtere komplekse og skiftende krav i tildelte prosjekter. Ved å implementere smidige praksiser.

# Det forventes at [virksomhetsnavn] integrerer smidige prinsipper.

# Proaktiv tilnærming til å implementere smidig arbeidsmetodikk i alle relevante prosjekter og oppgaver som omfattes av dette tildelingsbrevet.""",
]
# embed each of these, then take the average
inn_texts_embeddings = [embed(inn_text) for inn_text in inn_texts]
inn_texts_embeddings = np.array(inn_texts_embeddings)
inn_texts_embedding = np.mean(inn_texts_embeddings, axis=0)
inn_text_embedding = inn_texts_embedding
```

```python
populate_everything(df, inn_text_embedding)
```
